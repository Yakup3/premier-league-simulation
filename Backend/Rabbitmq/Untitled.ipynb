```python
import pandas as pd
```

```python
events_df = pd.read_csv("events.csv")
ginf_df = pd.read_csv("ginf.csv")
```

```python
ginf_df.head()
```

```python
premier_matches = ginf_df[(ginf_df["season"] == 2016) & (ginf_df["league"] == "E0")].reset_index(drop=True)
```

```python
events_df.head()
```

```python
goal_events = events_df[events_df["is_goal"] == 1]
```

```python
goals_at_premier = pd.merge(premier_matches, goal_events, left_on="id_odsp", right_on="id_odsp", how="left")
```

```python
goals_at_premier.head()
```

```python
goals_at_premier.columns
```

```python
columns = ["id_odsp", "date", "ht", "at", "id_event", "sort_order", "time", "text", "side", "event_team", "opponent", "player"]
goals_at_premier = goals_at_premier[columns]
```

```python
goals_at_premier["event_type"] = "goal"
```

```python
def add_match_start_end_and_event(grouped_df):
    df = grouped_df.copy()
    if df.shape[0] == 1 and pd.isna(df["id_event"].iloc[0]):
        row_as_dict = df.iloc[0].to_dict()
        df_data = [
            {
                **row_as_dict,
                "id_event": f'{row_as_dict.get("id_odsp", "bozuk/")[:-1]}0',
                "sort_order": 0,
                "time": 0,
                "event_type": "start"
            },
            {
                **row_as_dict,
                "id_event": f'{row_as_dict.get("id_odsp", "bozuk/")[:-1]}1',
                "sort_order": 1,
                "time": 90,
                "event_type": "finish"
            },
        ]
        df = pd.DataFrame(df_data)
        return df
    else:
        last_event = df["sort_order"].max()
        last_time = max(df["time"].max(), 90)

        first_row = df.iloc[0]

        match_start = first_row.copy()
        match_start["time"] = 0
        match_start["sort_order"] = 0
        match_start["event_type"] = "start"

        match_last = first_row.copy()
        match_last["time"] = last_time
        match_last["sort_order"] = last_event + 2
        match_last["event_type"] = "finish"

        df["sort_order"] = df["sort_order"].apply(lambda row: int(row + 1))
        df["id_event"] = df.apply(lambda row: row.get("id_odsp", "bozuk/")[:-1] + str(row["sort_order"]), axis=1)

        df.loc[-1] = match_last
        df.index = df.index + 1
        df.sort_index(inplace=True) 
        
        df.loc[-1] = match_start
        df.index = df.index + 1
        df = df.sort_values("sort_order").reset_index(drop=True)
        
        return df

goals_at_premier = goals_at_premier.groupby("id_odsp").apply(add_match_start_end_and_event).reset_index(drop=True)
```

```python
goals_at_premier["id_event"] = goals_at_premier.apply(lambda row: row.get("id_odsp")[:-1] + str(int(row["sort_order"])), axis=1)
```

```python
goals_at_premier = goals_at_premier["side"].fillna(-1)
```

```python
goals_at_premier
```

```python
int_columns = ["sort_order", "time", "side"]
for col in int_columns:
    goals_at_premier[col] = goals_at_premier[col].apply(int)
```
